```python
# Convert a file with the Interstial Force Constants from a text file into an array. This will allow the numerical values
# to be analyzed. 

%matplotlib inline
import numpy as np
from numpy import linalg as LA
import math
from matplotlib import pyplot


f=open("si.666.fc")     # open the file
Data = f.readlines()    # Read the file
Data = Data[17::]       # Discard the beginning of the file, which constains information on the lattice. It is not needed


# This does several things:
#   - it removes all the 'new line' characters with split
#   - it splits the individual numbers elements separated by a space into different list elements
for n in range(len(Data)):
    Data[n] = Data[n].strip().split()

    
    
# -------
# Separate and filter the list elements into an array 
# -------

# Create the array where the elements will be stored
IFC = np.zeros((len(Data),4))

# assign elements of the list into an array
for n in range(len(Data)):
    Data_temp = Data[n]
    for i in range(4):
        IFC[n,i] = Data_temp[i]


# Create a matrix that walks along the high symmetry lines. Gamma -> X -> K -> Gamma -> L. See the FCC 1st Brillouin zone
# (i.e. a BCC weignerseitz cell).

walk = np.zeros(shape=(36,3))

for i in range(0,10):
    walk[i,0] = i/10.

for i in range(11,20):
    walk[i,0] = 1.
    walk[i,1] = (i-10.)/10.

for i in range(21,30):
    walk[i,0] = (30.-i)/10.
    walk[i,1] = (30.-i)/10.
    
for i in range(31,36):
    walk[i,0] = (i-30.)/10.
    walk[i,1] = (i-30.)/10.
    walk[i,2] = (i-30.)/10.

# -------    
# Calculate the Fourier Transform along the high symmetry lines of the Interstitial Force Constants
# -------
C_q = np.zeros(shape=(36,36))

# The i is for different spatial and atomic positions. The k is for different q values. The l is along all the R values.
for i in range(36):
    for k in range(36):
        for l in range(216):
            # Complete the q and R dot product prior to the the fourier transform
            q_R = IFC[l+i*217+1,0]*walk[k,0]+IFC[l+i*217+1,1]*walk[k,1]+IFC[l+i*217+1,2]*walk[k,2]
            
            #C_q[k,i] += IFC[l+i*217+1,3]*abs(math.cos(q_R)-math.sin(q_R)*1j)   # Complete the Fourier Transform. Only take the real component.
            # The fortran code took an absolute value of the interstial force constants, so i thought i would try it.
            
            
            C_q[k,i] += IFC[l+i*217+1,3]*(math.cos(q_R))     # Complete the Fourier Transform

            
            
#-------
# Calculate the omega-squared and energy values by finding the eigenvalues of the Dynamical matrix
#-------
            
# The matrix above has the fourier transform of the Interstial Force Constants along the lines of Symmetry
# The Columns are the different spatial and atomic positions. There are 36 different values, which are specified by 4 different
# integers. The first two specify the direction of the movement of the 1st and 2nd atom, respectively (i.e. 1 is x, 2 is y,...).
# The second two integers specify, which atom is acting on, which atom. For example, 1 1 is the first atom on the first atom. 1 2
# is the first atom on the second atom. 

# For each q point row, we will build a 6x6 matrix, which represents the Dynamical matrix. We will find the 6 eigenvalues of
# this matrix, which correspond to the frequencies and energies at that q point.

# The Dynamical Matrix has the following structure:

#            ATOM 1       ATOM 2
#           _ x  y  z     x y z  _
#          |                     |
#        x |                     |
# ATOM 1 y |                     |
#        z |                     |
#          |                     |
#          |                     |
#        x |                     |
# ATOM 2 y |                     |
#        z |                     |
#          |                     |
#          |_                   _|


# Array to store the eigenvalues
E_v = np.zeros(shape=(6,36))

# Empty Dynamical Matrix. Will fill it in for each q point, find the eigenvalues, and then overwrite it on the next q point.
D_m = np.zeros(shape=(6,6))

# Matrices to store the 4 quadrants of the Dynamical Matrix. These will be used to build the dynamical matrix
a1a1 = np.zeros(9)
a1a2 = np.zeros(9)
a2a1 = np.zeros(9)
a2a2 = np.zeros(9)


for i in range(36):
    for j in range(0,36,4):
        a1a1[j/4] = C_q[i,j]
        a1a2[j/4] = C_q[i,j+1]
        a2a1[j/4] = C_q[i,j+2]
        a2a2[j/4] = C_q[i,j+3]
        
    for m in range(3):
        for n in range(3):
            D_m[m,n] = a1a1[m*3+n]
            D_m[m,n+3] = a1a2[m*3+n]
            D_m[m+3,n] = a2a1[m*3+n]
            D_m[m+3,n+3] = a2a2[m*3+n]
    
    A = LA.eigvals(D_m)
    for k in range(6):
        E_v[k,i] = A[k]
    
        
# Plot the bands
q = np.linspace(0.,35, 36)

for i in range(6):
    pyplot.plot(q, E_v[i,:], color='black')
    
    
```
